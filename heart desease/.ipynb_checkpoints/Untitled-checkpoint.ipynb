```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
heart = pd.read_csv('heart.csv')
```

```python
heart.head()
```

```python
heart.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
heart.groupby('HeartDisease')['HeartDisease'].count()
```

построенние матрицы рассеяния, которая покажет распределения внутри фрейма данных.

```python
colors = ['green','red']
cols = []
X = heart.loc[:, 'Age':'ST_Slope']
y=heart['HeartDisease']
for i in y:
    if i == 0:
        cols.append('green')
    else:
        cols.append('red')
grr = pd.plotting.scatter_matrix(X, figsize = (15,15), c=cols,marker = 'o', hist_kwds = {'bins':20}, s=60, alpha=.8)
```

```python
def categorical(df, cols):
    for col in cols:
        df[col] = df[col].astype('category')
    return df.info()
```

```python
categorical(X, ['Sex', 'ChestPainType','RestingECG','ST_Slope','ExerciseAngina'])
```

```python
X.head()
```

```python
X_dum = pd.get_dummies(X, drop_first=True)
```

```python
X_dum.head()
```

```python
from scipy.cluster.vq import whiten
scaled_X = whiten(X_dum)
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, random_state = 13)
```

после обработки данных, пробуем строить первую модель. Пробуем работать с к - ближайших соседей. 
Для обучения сперва попробуем подобрать параметры используя стандартный инструментарий языка. Цикл for

```python
from sklearn.neighbors import KNeighborsClassifier 
scores = []
n = np.arange(1,12)
for i in n:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    score = np.mean(y_pred == y_test)
    scores.append((i,score))
```

```python
np.sort(scores)
```

можно заметить, что наилучший рещультат достигается при работе с 9 ближайшими соседями, результать примерно 88%

Теперь используем метод  GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
params = {'n_neighbors': np.arange(1,15), 'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'] }
```

```python
knn_cv = GridSearchCV(knn, params, cv=10)
knn_cv.fit(X_train, y_train)
```

```python
knn_cv.best_score_
```

```python
knn_cv.best_params_
```

из перекрестной проверки видим, что при перекрестной проверке лучшим параметром оказалось число 12 с результатом в 86%

```python
knn_fin = KNeighborsClassifier(n_neighbors=12)
knn_fin.fit(X_train,y_train)
y_pred_fin = knn_fin.predict(X_test)
np.mean(y_pred_fin == y_test)
```

```python
knn_fin.score(X_test,y_test)
```

```python
import xgboost as xgb
```

```python
matrix = xgb.DMatrix(data = X_dum, label = y)
params = {"objective":"reg:logistic"}
max_depths = [1,10,20,30,40,50]
num_boost_round = [10,20, 30,40]
results = []
```

```python
for depth in num_boost_round:
    for depth in max_depth:
        params['max_depth']=depth
        results_cs = xgb.cv(dtrain=matrix, params=params, 
                  nfold=3, num_boost_round=100, early_stopping_rounds=100,
                  metrics="error", as_pandas=True, seed=13)
        results.append(results_cs['test-error-mean'])
np.min(results_cs["test-error-mean"])
```
