```python
reg_model = xgb.XGBRegressor()
gbm_param_grid = {
    'learning_rate': np.arange(0.05, 1, 0.05),
    'max_depth': np.arange(3, 100, 10),
    'n_estimators': np.arange(50, 200, 50),
    'gamma': np.arange(0.001, 2, 0.5),
    'colsample_bytree': np.arange(.1,1.05,.05),
    'subsample':np.arange(.1,.9,.03)
}

```

```python
chol_zero_df = df[df.Cholesterol == 0]
chol_zero_df = chol_zero_df.drop(columns = ['Cholesterol'])
pred_df = pd.get_dummies(chol_zero_df, drop_first=True)
pred_df.head()
```

```python
train_df = df[df.Cholesterol != 0]
y = train_df.Cholesterol
train_df = train_df.drop(['Cholesterol'], axis = 1)
train_df.head()
```

```python
train_dum = pd.get_dummies(train_df,drop_first=True)
train_dum.info()
```

```python
randomized_mean = RandomizedSearchCV(estimator =reg_model, param_distributions =gbm_param_grid, n_iter = 2, scoring = 'neg_mean_squared_error', verbose = 1,   cv = 10)
randomized_mean.fit(train_dum,y)
```

```python
test = randomized_mean.best_estimator_
test
```

```python
train_dum['Cholesterol'] = y
full = pd.concat([train_dum, pred_df])
full.head()
```

```python
y = full['HeartDisease']
```

```python
full = full.drop(['HeartDisease'], axis = 1)
```

```python
X = full
X.head()
```

```python
sns.scatterplot(data = full, x ='Age', y = 'Cholesterol', hue = y, alpha = 0.5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
pipeline = Pipeline([
                     ('scaler', StandardScaler()),
                     ("clf", xgb.XGBClassifier())
                    ])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=123)
```

```python
cross_val_scores = cross_val_score(pipeline, X, y, scoring="roc_auc", cv=3)

# Print avg. AUC
print("3-fold AUC: ", np.mean(cross_val_scores))
```

```python
gbm_param_grid = {
    'clf__learning_rate': np.arange(0.05, 1, 0.05),
    'clf__max_depth': np.arange(3, 10, 1),
    'clf__n_estimators': np.arange(50, 200, 50),
    'clf__gamma': np.arange(0.001, 2, 0.5)
}
```

```python
randomized_roc_auc = RandomizedSearchCV(estimator =pipeline, param_distributions =gbm_param_grid, n_iter = 25, scoring = 'roc_auc', verbose = 1,   cv = 6)
randomized_roc_auc.fit(X,y)
```

```python
print(randomized_roc_auc.best_score_)
print(randomized_roc_auc.best_estimator_)
```
